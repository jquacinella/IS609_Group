## First Attempt at Modeling Follow Relationships

Lets try to model the follow relationships, which is an important structure, on Twitter.

```python
import networkx as nx
g = nx.Graph()
g.add_node(0, {"handle": "@mrquntopolous"})
g.add_node(1, {"handle": "@SebastianThrun"})
g.add_edge(0, 1)

node_labels = nx.get_node_attributes(g,'handle')

plt.figure(figsize=(6,6))

pos = nx.spring_layout(g)
nx.draw(g, pos, arrows=True, node_size=900)
nx.draw_networkx_labels(g, pos, labels = node_labels)
```

## Issue with Graph Model From Class

The edge between two users indicating a follow relationship is a bit off. On Twitter, a follow relationship is not bi-directional. That is, if user A follows user B, it does not follow that user B follows user A. This is unlike other social networks, like Facebook, where friendships by definition are two way. How can we extend our model to account for this?

## Extending Graph Modeling

One way we can extend this model is to make edges directional. In the graph models from class, edges just connect two nodes with no presumption of direction. In directed-graphs (http://en.wikipedia.org/wiki/Directed_graph), each edge has a direction. This means that one node is the 'from' node while the other is the 'to' node. In diagrams, this is indicated by the arrow head on the edge.

This will be useful for our follow relationships. Lets extend our model by using NetworkX's DiGraph model, which allows for directed edges:

```python
g = nx.DiGraph()
g.add_node(0, {"handle": "@mrquntopolous"})
g.add_node(1, {"handle": "@SebastianThrun"})
g.add_edge(0, 1)

node_labels = nx.get_node_attributes(g,'handle')

plt.figure(figsize=(6,6))

pos = nx.spring_layout(g)
nx.draw(g, pos, arrows=True, node_size=900)
nx.draw_networkx_labels(g, pos, labels = node_labels)
```

Now we can see that it is @mrquintopolous who follows @SebastianThrun, and not the other way around.

## Modeling Tweets, Favorites and Re-Tweets

What if we need to model the tweets that user send out themselves, and how users interact with them? However, we have another problem: the previous nodes represented users, so how will we model tweets or other entities? In the graph models we have used prior, typically a node represented one kind of thing, in our case users. How can we model more than one kind of entity?

Using a [property graph](https://github.com/tinkerpop/gremlin/wiki/Defining-a-Property-Graph), we can add properties to nodes and edges, which can help us differentiate between different kinds of enetities and relationships in our graphs. For example, we can add a node that represents a tweet made by one of these users. We will color it differently to emphasize that its a different kind of node.

```python
g = nx.DiGraph()
g.add_node(0, {"id": "@mrquntopolous", "color": "r"})
g.add_node(1, {"id": "@SebastianThrun", "color": "r"})
g.add_node(2, {"id": "Tweet #530038648860065793", "color": "orange"})
g.add_edge(0, 1, {"type": "follows"})
g.add_edge(0, 2, {"type": "tweeted"})

node_labels = nx.get_node_attributes(g,'id')
edge_labels = nx.get_edge_attributes(g,'type')
colors = nx.get_node_attributes(g, 'color')

plt.figure(figsize=(6,6))

pos = nx.spring_layout(g)
nx.draw(g, pos, arrows=True, node_size=900, node_color=colors.values())
nx.draw_networkx_labels(g, pos, labels = node_labels)
nx.draw_networkx_edge_labels(g, pos, labels = edge_labels)
```

Here, each node type has a different color: red for Twitter users, orange for tweets themselves. We have also given edge's a property, called 'type'. This property tells you what kind of relationship it indicates: in this example, we have a 'follows' relationship and a 'tweeted' relationship. Notice: it only makes sense to connect two users via an edge of type 'follows'; same for edges of type 'tweeted', which only make sense connectig users and tweets.
