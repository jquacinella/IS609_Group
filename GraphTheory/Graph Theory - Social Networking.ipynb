## First Attempt at Modeling Follow Relationships

Lets try to model the follow relationships, which is an important structure, on Twitter.

```python
import networkx as nx
g = nx.Graph()
g.add_node(0, {"handle": "@mrquntopolous"})
g.add_node(1, {"handle": "@SebastianThrun"})
g.add_edge(0, 1)

node_labels = nx.get_node_attributes(g,'handle')

plt.figure(figsize=(6,6))

pos = nx.spring_layout(g)
nx.draw(g, pos, arrows=True, node_size=900)
nx.draw_networkx_labels(g, pos, labels = node_labels)
```

## Issue with Graph Model From Class

The edge between two users indicating a follow relationship is a bit off. On Twitter, a follow relationship is not bi-directional. That is, if user A follows user B, it does not follow that user B follows user A. This is unlike other social networks, like Facebook, where friendships by definition are two way. How can we extend our model to account for this?

## Extending Graph Modeling

One way we can extend this model is to make edges directional. In the graph models from class, edges just connect two nodes with no presumption of direction. In directed-graphs (http://en.wikipedia.org/wiki/Directed_graph), each edge has a direction. This means that one node is the 'from' node while the other is the 'to' node. In diagrams, this is indicated by the arrow head on the edge.

This will be useful for our follow relationships. Lets extend our model by using NetworkX's DiGraph model, which allows for directed edges:

```python
g = nx.DiGraph()
g.add_node(0, {"handle": "@mrquntopolous"})
g.add_node(1, {"handle": "@SebastianThrun"})
g.add_edge(0, 1)

node_labels = nx.get_node_attributes(g,'handle')

plt.figure(figsize=(6,6))

pos = nx.spring_layout(g)
nx.draw(g, pos, arrows=True, node_size=900)
nx.draw_networkx_labels(g, pos, labels = node_labels)
```

Now we can see that it is @mrquintopolous who follows @SebastianThrun, and not the other way around.

## Modeling Tweets, Favorites and Re-Tweets

What if we need to model the tweets that user send out themselves, and how users interact with them? However, we have another problem: the previous nodes represented users, so how will we model tweets or other entities? In the graph models we have used prior, typically a node represented one kind of thing, in our case users. How can we model more than one kind of entity?

Using a [property graph](https://github.com/tinkerpop/gremlin/wiki/Defining-a-Property-Graph), we can add properties to nodes and edges, which can help us differentiate between different kinds of enetities and relationships in our graphs. For example, we can add a node that represents a tweet made by one of these users. We will color it differently to emphasize that its a different kind of node.

```python
g = nx.DiGraph()
g.add_node(0, {"id": "@mrquntopolous", "color": "r"})
g.add_node(1, {"id": "@SebastianThrun", "color": "r"})
g.add_node(2, {"id": "Tweet #530038648860065793", "color": "orange"})
g.add_edge(0, 1, {"type": "follows"})
g.add_edge(0, 2, {"type": "tweeted"})

node_labels = nx.get_node_attributes(g,'id')
edge_labels = nx.get_edge_attributes(g,'type')
colors = nx.get_node_attributes(g, 'color')

plt.figure(figsize=(6,6))

pos = nx.spring_layout(g)
nx.draw(g, pos, arrows=True, node_size=900, node_color=colors.values())
nx.draw_networkx_labels(g, pos, labels = node_labels)
nx.draw_networkx_edge_labels(g, pos, labels = edge_labels)
```

Here, each node type has a different color: red for Twitter users, orange for tweets themselves. We have also given edge's a property, called 'type'. This property tells you what kind of relationship it indicates: in this example, we have a 'follows' relationship and a 'tweeted' relationship. Notice: it only makes sense to connect two users via an edge of type 'follows'; same for edges of type 'tweeted', which only make sense connectig users and tweets.

# Graph Centrality

In graph theory, [*centrality*](http://en.wikipedia.org/wiki/Centrality) refers to various metrics which identify which of the vertices within a graph are the most important. For our purposes, we would like to identify the most influential people in the data science community that we crawled. There are many different ways of defining centrality, including:

* Degree Centrality - One metric that can help define which vertices are important is using the number of links incident upon a node. The more edges on a node, the more 'important' it might be. This is one the simplest metrics for centrality. 

* Closeness Centrality - In graph theory, the shortest path is the minimal amount of hops needed to travel between two nodes. We define another distance metric that is defined as the sum of its shortest path distances to all other nodes. The smaller the number, the closer this node is to all other nodes, so we can define closeness as the reciprocal of this metric. We will not use this metric.

* Betweenness Centrality - this metric quantifies the number of times a node acts as a 'bridge' along the shortest path between two other nodes. The more often a node is along shortest path in the graph, the more likely it is to be important. Notice how both closeness and betweenness are defined in terms of the whole graph, while degree centrality is defined locally per node.

* PageRank - is an iterative algorithm that outputs a probability distribution over nodes that represents the likelihood that a person randomly moving along edges will arrive at any particular node. After each iteration, the node's value is an approximate PageRank value that approaches the theoretical true value over time.


## Degree Centrality and Degree Distribution

Given the above definition of degree centrality, lets look at our crawled graph and see if we can find nodes with a high number of edges coming in. Using networkx's in_degree() function, we can calculate every node's in degree. Using this, lets look at a histogram of in-degree values:

```python
degree = g.in_degree()
in_degrees = np.asarray(degree.values())
in_degrees.sort()

hist, bins = np.histogram(in_degrees, bins=20)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2

fig = plt.figure()
fig.suptitle('In Degree Frequency Distribution', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
ax.set_xlabel('In Degree')
ax.set_ylabel('Frequency')

plt.bar(center, hist, align='center', width=width)
plt.show()
```

## Page Rank

PageRank is an interative algorithm that outputs a probability distribution over nodes that represents the likelihood that a person randomly moving along edges will arrive at any particular node. After each iteration, the node's value is an approximate PageRank value that approaches the theoretical true value over time. PageRank is at the core of Google's web crawling. 

Using NetworkX, we can use this algorithm to see what PageRank thinks are the most 'relevant' nodes, and compare them to what we found using Degree centrality:

```python
# Get PageRank values, and sort them by value
import operator
pr = nx.pagerank(g, alpha=0.9)
sorted_pr = sorted(pr.items(), key=operator.itemgetter(1))

# Print top 20
for node in sorted_pr[-20:]:
    print node[1], user_dict[node[0]]['name']
```

As we can see, there is a lot of overlap, but there are some new nodes, like Kara Swisher, Baratunde, John Maeda and Joshua Schachter. However, these people would not be considered in the Data Science community. This shows that PageRank finds 'relevant' or 'popular' nodes, which includes users who happen to be popular in other communities.

# Extending the Model with Tweets

Another question we would like to ask of the graph is "In terms of tweet propagation, can we measure a user's influence?" Services like Klout try to measure how influential a user is. One factor in determining influence by seeing how their tweets are propagated to others, or what is known as *reach*. We can try to model this using information we can get about tweets in the API. TO illustrate, the graph can be extended to have some nodes presenting a tweet, with edges from users to tweets representing the relationship "has retweeted". 

Given a tweet, we can find the users who retweeted it their followers. However, crawling all these users into a single graph proves problematic, memory-wise, but we will have to estimate this.

If we count the number of followers, we can get an upper-estimate to the number of users who saw this tweet on their timelimes.

```python
import time

tweet_id=541746341295452160

followers = { }
followers_count = 0

users = api.GetRetweeters(tweet_id)
for user in users:
    while True:
        try:
            print "User %d" % user
            followers[user] = api.GetFollowerIDs(user_id=user)
            break
        except:
            print "Sleeping ..."
            time.sleep(60*5)
            continue
    
    followers_count += len(followers[user])
    print followers_count

#pickle.dump(followers, open("followers", "wb"), pickle.HIGHEST_PROTOCOL)
```
